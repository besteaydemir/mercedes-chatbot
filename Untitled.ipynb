```python
def get_power(table):
    value_dict = {
        'power': None
    }
    rows = table.find_all('tr')
    value_dict['power'] = rows[0].get_text().strip().replace('\n','-')
    return value_dict
```

```python
def get_performance_info(table):
    value_dict = {
        'range': None,
        'top_speed': None
    }
    rows = table.find_all('tr')
    value_dict['top_speed'] = rows[1].get_text().strip().replace('\n','-')
    value_dict['range'] = rows[2].get_text().strip().replace('\n','-')
    return value_dict
```

```python
def get_battery(table):
    value_dict = {
        'warranty': None,
        'type': None,
        'capacity': None
    }
    rows_1 = table[0].find_all('tr')
    rows_2 = table[1].find_all('tr')
    value_dict['type'] = rows_1[1].get_text().strip().replace('\n','-')
    value_dict['warranty'] = rows_1[4].get_text().strip().replace('\n','-')
    value_dict['capacity'] = rows_2[0].get_text().strip().replace('\n','-')
    return value_dict
```

```python
def get_price(table):
    value_dict = {
        'price': None
    }
    rows = table.find_all('tr')
    value_dict['price'] = " - ".join(rows[2].get_text().split())
    return value_dict
```

```python
def get_charge_info(table):
    value_dict = {
        'charge_time': None,
        'charge_speed': None
    }
    rows = table[1].find_all('tr')
    value_dict['charge_time'] = " - ".join(rows[0].get_text().split())
    value_dict['charge_speed'] = " - ".join(rows[1].get_text().split())
    return value_dict
```

```python
def get_seater_info(soup):
    value_dict = {
        'seater': None,
        'body': None,
        'segment': None
    }
    data_tables = soup.find_all('div', class_='data-table')
    seater = None
    for data_table in data_tables:
        h2_tag = data_table.find('h2', text='Miscellaneous')
        if h2_tag:
            tables = data_table.find_all('table')
            if tables:
                value_dict['seater'] = tables[0].find_all('tr')[0].find_all('td')[1].get_text()
                value_dict['body'] = tables[1].find_all('tr')[0].find_all('td')[1].get_text()
                value_dict['segment'] = tables[1].find_all('tr')[1].find_all('td')[1].get_text()
    return value_dict
```

```python
def get_car_details(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    car_deets = dict()
    all_tables = soup.find_all('table')
    car_deets['price'] = get_price(all_tables[0])
    car_deets['performance'] = get_performance_info(all_tables[4])
    car_deets['power'] = get_power(all_tables[5])
    car_deets['battery'] = get_battery(all_tables[6:8])
    car_deets['charge'] = get_charge_info(all_tables[8:10])
    car_deets['misc'] = get_seater_info(soup)  
    return car_deets
    
    
```

```python
cars = {
    'EQS 450+': 'https://ev-database.org/imp/car/2008/Mercedes-Benz-EQS-450plus',
    'EQE 500 4Matic': 'https://ev-database.org/imp/car/1900/Mercedes-Benz-EQE-500-4MATIC',
    'G53 4Matic+': 'https://ev-database.org/imp/car/2012/Mercedes-Benz-EQS-AMG-53-4MATICplus',
    'EQS SUV 450+': 'https://ev-database.org/imp/car/2087/Mercedes-Benz-EQS-SUV-450plus',
    'EQS 580 4Matic': 'https://ev-database.org/imp/car/2011/Mercedes-Benz-EQS-580-4MATIC',
    'EQE 300': 'https://ev-database.org/imp/car/1897/Mercedes-Benz-EQE-300',
    'EQS SUV Maybach 680': 'https://ev-database.org/imp/car/2086/Mercedes-Benz-EQS-SUV-Maybach-680',
    'EQB 250+': 'https://ev-database.org/imp/car/1988/Mercedes-Benz-EQB-250plus',
    'eVito Tourer': 'https://ev-database.org/imp/car/2140/Mercedes-Benz-eVito-Tourer-Long-90-kWh'
}
```

```python
car_info = dict()
```

```python
for car_name in tqdm(cars.keys()):
    car_info[car_name] = get_car_details(cars[car_name])
```

```python
car_prices = list()
car_range = list()
car_top_speed = list()
car_power = list()
car_battery_warranty = list()
car_battery_type = list()
car_battery_capacity = list()
car_charge_time, car_charge_speed, car_seater, car_body, car_segment = list(), list(), list(), list(), list()
```

```python
for car_name in tqdm(cars.keys()):
    spec_car_info = car_info[car_name]
    car_prices.append(spec_car_info['price']['price'])
    car_range.append(spec_car_info['performance']['range'])
    car_top_speed.append(spec_car_info['performance']['top_speed'])
    car_power.append(spec_car_info['power']['power'])
    car_battery_warranty.append(spec_car_info['battery']['warranty'])
    car_battery_type.append(spec_car_info['battery']['type'])
    car_battery_capacity.append(spec_car_info['battery']['capacity'])
    car_charge_time.append(spec_car_info['charge']['charge_time'])
    car_charge_speed.append(spec_car_info['charge']['charge_speed'])
    car_seater.append(spec_car_info['misc']['seater'])
    car_body.append(spec_car_info['misc']['body'])
    car_segment.append(spec_car_info['misc']['segment'])
```

```python
df = pd.DataFrame({
    'Name': cars.keys(),
    'Price': car_prices,
    'Range': car_range,
    'Top Speed': car_top_speed,
    'Power': car_power,
    'Battery Warranty': car_battery_warranty,
    'Battery Type': car_battery_type,
    'Battery Capacity': car_battery_capacity,
    'Charge Time': car_charge_time,
    'Charge Speed': car_charge_speed,
    'Seater': car_seater,
    'Body': car_body,
    'Segment': car_segment
})
```

```python
df.Price = df.Price.apply(lambda x: float(x.split('-')[1].replace('€', '').replace(',', '')))
df.Range = df.Range.apply(lambda x: float(x.split('-')[1].split(' ')[0]))
df['Top Speed'] = df['Top Speed'].apply(lambda x: float(x.split('-')[1].split(' ')[0]))
df['Power'] = df['Power'].apply(lambda x: float(x.split('(')[1].split(' ')[0]))
df['Battery Capacity'] = df['Battery Capacity'].apply(lambda x: float(x.split('-')[1].split(' ')[0]))
df['Charge Time'] = df['Charge Time'].apply(lambda x: x.split('-')[5])
df['Charge Speed'] = df['Charge Speed'].apply(lambda x: float(x.split('-')[2]))
df['Seater'] = df['Seater'].apply(lambda x: float(x.split(' ')[0]))
```

```python
df.to_csv('vehicle_data.csv', index=False)
```

```python
def format_car_data(car_data, indent=0):
    formatted_text = ''
    for key, value in car_data.items():
        if isinstance(value, dict):
            formatted_text += f"{' ' * indent}{key}:\n"
            formatted_text += format_car_data(value, indent + 4)
        else:
            formatted_text += f"{' ' * indent}{key}: {value}\n"
    return formatted_text


formatted_text = format_car_data(car_info)
print(formatted_text)
```
